### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_='part_list_2').find_all('h3'):
    print(d.find(class_='date').text, d.em.text, d.a.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
datas = {}
for d in soup.find(class_='part_list_2').find_all('h3'):
    if d.em.text not in datas:
        datas[d.em.text] = []
    datas[d.em.text].append({'date': d.find(class_='date').text, 'title': d.a.text, 'url': d.a['href']})
        
print(datas)
```

### ③ 哪一個類別的文章最多

```python
max = 0
maxType = ''
for type in datas:
    if len(datas[type]) > max:
        max = len(datas[type])
        maxType = type
print("\"{0}\"類別有最多篇文章，共有 {1} 篇".format(maxType, max))
```
