```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

import requests
import json

def getData(url):
    response = requests.get(url)
    data = json.loads(response.text)
    return data

def getTotalCount(data):
    global totalCommentCount
    global totalLikeCount 
    for d in data:
        totalCommentCount += d['commentCount']
        totalLikeCount += d['likeCount']

    
#è¨ˆç®—ç†±é–€æ–‡ç« ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
totalCommentCount = 0
totalLikeCount = 0 

getTotalCount(getData('https://www.dcard.tw/_api/forums/pet/posts?popular=true'))
print(f"ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ç‚º{totalCommentCount/len(data)}ï¼Œå¹³å‡æŒ‰è®šäººæ•¸ç‚º{totalLikeCount/len(data)}")

#è¨ˆç®—éç†±é–€æ–‡ç« ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
totalCommentCount = 0
totalLikeCount = 0 

getTotalCount(getData('https://www.dcard.tw/_api/forums/pet/posts?popular=false'))
print(f"éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ç‚º{totalCommentCount/len(data)}ï¼Œå¹³å‡æŒ‰è®šäººæ•¸ç‚º{totalLikeCount/len(data)}")
```
