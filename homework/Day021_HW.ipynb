## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
# 匯入套件
import requests
import time
import re
from datetime import datetime
from bs4 import BeautifulSoup
from selenium import webdriver


today = datetime.strptime(datetime.now().strftime('%Y/%m/%d') ,'%Y/%m/%d')
length = 0

# 以Chromedriver開啟網頁
with webdriver.Chrome(executable_path = './chromedriver') as browser:
    # 載入指定網頁
    browser.get('https://www.ettoday.net/news/news-list.htm')
    
    
    while True:
        # 解析html
        html_source = browser.page_source
        soup = BeautifulSoup(html_source)
        
        # 取出最後一筆資料的時間
        lasttime = soup.find(class_ = 'part_list_2').find_all(name = 'h3')[-1].span.text
        
        # 轉換成時間型態
        lasttime = datetime.strptime(lasttime, '%Y/%m/%d %H:%M')
        
        # 判斷最後一筆資料時間是否仍在今天
        if lasttime < today:
            for d in soup.find(class_ = 'part_list_2').find_all(name = 'h3'):
                if d.find(name = 'span', text = re.compile(datetime.now().strftime('%Y/%m/%d'))) == None:
                    continue
                print(d.span.text, d.em.text, d.a.text)
            break
        else:
            if length != len(soup.find(class_ = 'part_list_2').find_all(name = 'h3')):
                length = len(soup.find(class_ = 'part_list_2').find_all(name = 'h3'))
            else:
                for d in soup.find(class_ = 'part_list_2').find_all(name = 'h3'):
                    if d.find(name = 'span', text = re.compile(datetime.now().strftime('%Y/%m/%d'))) == None:
                        continue
                    print(d.span.text, d.em.text, d.a.text)
                break
        
        # 滾動頁面
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(0.5)
        

```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''

# 匯入套件
import requests
import time
import re
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from selenium import webdriver


today = datetime.strptime(datetime.now().strftime('%Y/%m/%d') ,'%Y/%m/%d')
threedaysago = today - timedelta(days = 3)

# 以Chromedriver開啟網頁
with webdriver.Chrome(executable_path = './chromedriver') as browser:
    # 載入指定網頁
    browser.get('https://www.ettoday.net/news/news-list-%s-0.htm' %(datetime.strftime(threedaysago, '%Y-%m-%d')))
    
    
    while True:
        # 解析html
        html_source = browser.page_source
        soup = BeautifulSoup(html_source)
        
        # 取出最後一筆資料的時間
        lasttime = soup.find(class_ = 'part_list_2').find_all(name = 'h3')[-1].span.text
        
        # 轉換成時間型態
        lasttime = datetime.strptime(lasttime, '%Y/%m/%d %H:%M')
        
        # 判斷最後一筆資料時間是否仍在今天
        if lasttime < threedaysago + timedelta(hours = 15):
            for d in soup.find(class_ = 'part_list_2').find_all(name = 'h3'):
                if d.find(name = 'span', text = re.compile(threedaysago.strftime('%Y/%m/%d') + '\s(15:\d\d)|(16:\d\d)|(17:00)')) == None:
                    continue
                print(d.span.text, d.em.text, d.a.text)
            break
        
        # 滾動頁面
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(0.5)

```
