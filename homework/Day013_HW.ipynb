## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'

resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'html5lib')

for d in soup.find_all(class_='title'):
    print('-----------------------------------------------------')
    print('[Title]: ' ,d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc' + d.a['href']).text, 'html5lib')
        print('[Author]:' ,r.find_all(class_='article-meta-value')[0].text)
        print('[Time]:  ' ,r.find_all(class_='article-meta-value')[3].text)
    except:
        continue
```

### ③ 試著爬爬看其他版的文章

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/Gossiping/index.html'


resp = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(resp.text, 'html5lib')

for d in soup.find_all(class_='title'):
    print('-----------------------------------------------------')
    print('[Title]: ' ,d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc' + d.a['href'], cookies={'over18': '1'}).text, 'html5lib')
        print('[Author]:' ,r.find_all(class_='article-meta-value')[0].text)
        print('[Time]:  ' ,r.find_all(class_='article-meta-value')[3].text)
    except:
        continue
print('-----------------------------------------------------')
```
